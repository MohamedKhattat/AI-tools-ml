````python
import pandas as pd
import pdfplumber
from PIL import Image

def open_pdf(file_path):
    """Open a PDF file and return the pdf object."""
    try:
        pdf = pdfplumber.open(file_path)
        return pdf
    except Exception as e:
        print(f"Error opening PDF: {e}")
        return None

def extract_tables_from_page(pdf, page_num):
    """Extract tables from a specified page."""
    try:
        page = pdf.pages[page_num]
        tables = page.extract_tables()
        return tables
    except IndexError:
        print(f"Page {page_num} not found.")
        return []
    except Exception as e:
        print(f"Error extracting tables: {e}")
        return []

def process_tables(tables):
    """Process extracted tables into DataFrames."""
    dfs = []
    for i, table in enumerate(tables):
        df = pd.DataFrame(table[1:], columns=table[0])
        # Optional cleaning of "Group" column
        if "Group" in df.columns:
            df["Group"] = df["Group"].str.replace(" ", "")
        dfs.append(df)
    return dfs

def display_image_from_page(page, save_path="page_image.png"):
    """Convert a page to an image and display it."""
    page_image = page.to_image()
    page_image.save(save_path)
    img = Image.open(save_path)
    img.show()

````

## Testing the Output

This section describes how the notebook handles the output of extracted tables based on the number of tables present in the specified PDF page.

### Handling Single and Multiple Tables

The extraction process can result in one or more tables being found on the page. To effectively manage the output, the code implements the following logic:

- **Single Table Case**: 
  - If the PDF page contains only one table, the DataFrame representing that table can be accessed directly using `dataframes[0]`. 
  - This means that the output will display only the first (and only) DataFrame without any need for iteration.

- **Multiple Tables Case**: 
  - If the PDF page contains multiple tables, the DataFrames can be accessed using an index ranging from `dataframes[0]` to `dataframes[n-1]`, where `n` is the total number of tables extracted.
  - The code iterates through the `dataframes` list, printing each DataFrame sequentially, which ensures all extracted tables are displayed clearly.

### Example Output Logic

Here's a simplified representation of the output logic:

```python
if dataframes:

## Example 1
This section describes how the notebook handles the output of extracted tables for the third page on declaration document which contain only one table



````python
# Example usage
pdf_file = 'declaration.pdf'
pdf = open_pdf(pdf_file)

if pdf:
    page_num = 2  # Example page number
    tables = extract_tables_from_page(pdf, page_num)
    
    if tables:
        dataframes = process_tables(tables)
        for i, df in enumerate(dataframes):
            print(f"\nTable {i + 1}:")
            print(df)
    else:
        print("No tables found on the specified page.")

    # Close the PDF
    pdf.close()
````

````python
# Example usage
pdf_file = 'declaration.pdf'
pdf = open_pdf(pdf_file)

if pdf:
    page_num = 10  # Example page number
    tables = extract_tables_from_page(pdf, page_num)
    
    if tables:
        dataframes = process_tables(tables)
        if dataframes:  # Ensure there are DataFrames to display
            for i, df in enumerate(dataframes):
                print(f"Table {i + 1}:")
                print(df)
        else:
            print("No DataFrames created from the extracted tables.")
    else:
        print("No tables found on the specified page.")

    # Close the PDF
    pdf.close()
````
