```python
import pdfplumber
import pandas as pd
from PIL import Image

# Open the PDF
pdf = pdfplumber.open('declaration.pdf')

# Extract the first page (page 7 in this case)
p0 = pdf.pages[6]

# Convert the PDF page to an image (optional, if rendering is supported)
p0_image = p0.to_image()

# Save the page as an image (optional)
p0_image.save("p0_image.png")

# Open the saved image with PIL
img = Image.open("p0_image.png")

# Display the image (if in an interactive environment)
img.show()

# Extract tables (using extract_tables to get multiple tables)
tables = p0.extract_tables()

# Check if there are multiple tables
if len(tables) >= 2:
    # First table
    table_1 = tables[0]
    df1 = pd.DataFrame(table_1[1:], columns=table_1[0])

    # Second table (assuming the second one has different columns)
    table_2 = tables[1]
    df2 = pd.DataFrame(table_2[1:], columns=["Group", "Line", "Amount at Beginning of Year", "Amount at End of Year"])

    # Optional: clean up the "Group" column in df2
    if "Group" in df2.columns:
        df2["Group"] = df2["Group"].str.replace(" ", "")
    else:
        print("'Group' column not found in the second table")

    # Close the PDF
    pdf.close()

    # Display the results
    print("First Table:")
    print(df1)
    print("\nSecond Table:")
    print(df2)
else:
    print("Unable to find two tables on the page.")

# Close the PDF (in case not done earlier)
pdf.close()
```

```python
import pdfplumber
import pandas as pd
from PIL import Image

# Open the PDF
pdf = pdfplumber.open('declaration.pdf')

# Extract the specific page (page 10 in this case)
p0 = pdf.pages[11]

# Convert the PDF page to an image (optional, for visualization)
p0_image = p0.to_image()

# Save the page as an image (optional)
p0_image.save("p0_image.png")

# Open the saved image with PIL (optional)
img = Image.open("p0_image.png")

# Display the image (if in an interactive environment)
img.show()

# Extract tables from the page
tables = p0.extract_tables()

# Check if there are at least two tables on the page
if len(tables) >= 2:
    # First table
    table_1 = tables[0]
    df1 = pd.DataFrame(table_1[1:], columns=table_1[0])  # Automatically use extracted columns
    
    # Second table
    table_2 = tables[1]
    df2 = pd.DataFrame(table_2[1:], columns=table_2[0])  # Automatically use extracted columns

    # Ensure both tables have "Group" and "Line" columns, if present
    for df, table_num in zip([df1, df2], ["First", "Second"]):
        if "Group" in df.columns and "Line" in df.columns:
            df["Group"] = df["Group"].str.replace(" ", "")  # Clean up the "Group" column
        else:
            print(f"'Group' and 'Line' columns not found in the {table_num} table")

    # Close the PDF
    pdf.close()

    # Display the results
    print("First Table:")
    print(df1)
    print("\nSecond Table:")
    print(df2)
else:
    print("Unable to find two tables on the page.")

# Close the PDF (in case not done earlier)
pdf.close()
```

```python
import pdfplumber
import pandas as pd
from PIL import Image

# Open the PDF
pdf = pdfplumber.open('declaration.pdf')

# Extract the specific page (page 10 in this case)
p0 = pdf.pages[14]

# Convert the PDF page to an image (optional, for visualization)
p0_image = p0.to_image()

# Save the page as an image (optional)
p0_image.save("p0_image.png")

# Open the saved image with PIL (optional)
img = Image.open("p0_image.png")

# Display the image (if in an interactive environment)
img.show()

# Extract tables from the page
tables = p0.extract_tables()

# Check if there are at least two tables on the page
if len(tables) >= 2:
    # First table (structure with 7 columns)
    table_1 = tables[0]
    df1 = pd.DataFrame(table_1[1:], columns=table_1[0])  # Automatically use extracted columns
    
    # Second table (which includes "Full Name" and "Nationality")
    table_2 = tables[1]
    df2 = pd.DataFrame(table_2[1:], columns=table_2[0])  # Automatically use extracted columns

    # Ensure both tables are processed accordingly

    # For the first table: check if "Group" and "Line" columns exist
    if "Group" in df1.columns and "Line" in df1.columns:
        df1["Group"] = df1["Group"].str.replace(" ", "")  # Clean up the "Group" column
    else:
        print("'Group' and 'Line' columns not found in the First table")

    # For the second table: handle "Full Name" and "Nationality"
    if "Full Name" in df2.columns and "Nationality" in df2.columns:
        # Clean up if necessary
        df2["Full Name"] = df2["Full Name"].str.strip()  # Clean up spaces from "Full Name"
        df2["Nationality"] = df2["Nationality"].str.strip()  # Clean up spaces from "Nationality"
    else:
        print("'Full Name' and 'Nationality' columns not found in the Second table")

    # Close the PDF
    pdf.close()

    # Display the results
    print("First Table:")
    print(df1)
    print("\nSecond Table:")
    print(df2)
else:
    print("Unable to find two tables on the page.")

# Close the PDF (in case not done earlier)
pdf.close()
```
