```python
from pdf2image import convert_from_path
import pytesseract

# Path to the PDF file
pdf_path = r'C:/Users/DELL/Downloads/quittance3_(5)_(1)[2].pdf'

# Convert PDF to images
images = convert_from_path(pdf_path)

# Loop through each image (each page)
for i, image in enumerate(images):
    # Convert image to text using Tesseract OCR
    text = pytesseract.image_to_string(image, lang='eng+fra')
    
    # Print or save the extracted text
    print(f"Page {i+1}:\n{text}\n{'-'*50}")
```

```python
# Initialize EasyOCR Reader
reader = easyocr.Reader(['en'])
# Extract text from each image
for i, image in enumerate(images):
    result = reader.readtext(image)
    print(f"Page {i+1}:")
    for (bbox, text, prob) in result:
        print(f"Text: {text}\nConfidence: {prob}\n")
```

```python
import fitz  # PyMuPDF

pdf_path = r'C:/Users/DELL/Downloads/quittance3_(5)_(1)[2].pdf'

# Open the PDF file
pdf_document = fitz.open(pdf_path)

# Extract text
full_text = ""
for page_num in range(len(pdf_document)):
    page = pdf_document.load_page(page_num)
    full_text += page.get_text()

# Extract images
images = []
for page_num in range(len(pdf_document)):
    page = pdf_document.load_page(page_num)
    for img_index, img in enumerate(page.get_images(full=True)):
        xref = img[0]
        base_image = pdf_document.extract_image(xref)
        image_bytes = base_image["image"]
        image_filename = f"page_{page_num+1}_image_{img_index+1}.png"
        with open(image_filename, "wb") as image_file:
            image_file.write(image_bytes)
        images.append(image_filename)

```

```python
# Extract metadata
metadata = pdf_document.metadata

# Close the PDF document
pdf_document.close()

# Output results
print("Text Extracted:")
print(full_text[:1000])  
print("\nImages Extracted:")
for image in images:
    print(image)
print("\nMetadata Extracted:")
print(metadata)
```
